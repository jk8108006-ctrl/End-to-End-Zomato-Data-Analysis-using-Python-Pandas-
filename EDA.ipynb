```python
sns.set_theme(style="whitegrid", palette="muted")
print("=" * 50)
print("  BASIC DATASET INFO")
print("=" * 50)
print(f"Total restaurants : {len(df)}")
print(f"Total columns     : {df.shape[1]}")
print(f"Locations covered : {df["location"].nunique()}")
print(f"Cuisine types      : {df['primary_cuisine'].nunique()}")
print(f"Avg rating         : {df['rate'].mean():.2f}")
print(f"Avg cost for two   : ₹{df['approx_cost(for two people)'].mean():.0f}")
```

```python
online_analysis = df.groupby("online_order").agg(
    avg_rating = ("rate","mean"),
    avg_cost   = ("approx_cost(for two people)", "mean"),
    total_restaurants = ("name","count")).round(2)

print("\nOnline Order Analysis :")
print(online_analysis)
```

```python
top_location = (df.groupby('location')
                   .agg(
                       restaurants = ('name', 'count'),
                       avg_rating  = ('rate', 'mean'),
                       avg_cost    = ('approx_cost(for two people)', 'mean')
                   )
                   .round(2)
                   .sort_values('restaurants', ascending=False)
                   .head(10))

print("\nTop 10 Locations :")
print(top_location)
```

```python
price_rating = (df.groupby("price_category")["rate"]
                .mean()
                .round(2)
                .sort_values(ascending=False))

print("\nAvg Rating by Price Category :")
print(price_rating)
```

```python
pivot = pd.pivot_table(
    df,
    columns = "online_order",
    index   = "price_category",
    values  = "rate",
    aggfunc = "mean"
).round(2)

print("\nPivot — Avg Rating (Price × Online Order) :")
print(pivot)
```

```python
top_cuisines = df["primary_cuisine"].value_counts().head(8)
print("\nTop 8 Cuisines :")
print(top_cuisines)
```

```python
corr = df[['rate', 'votes', 
           'approx_cost(for two people)']].corr().round(3)
print("\nCorrelation Matrix :")
print(corr)
```
