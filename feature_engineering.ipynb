# Feature Engineering

Feature engineering is one of the most important concepts in data analysis and machine learning.
Feature engineering is the process of creating new features (columns) or transforming existing features to make the data more useful for analysis or machine learning.

```python
df['approx_cost(for two people)'].describe()
```

```python
def price_bucket(cost):
    if cost<=350:
        return "Budget"
    elif cost<=700:
        return "Mid-Range"
    elif cost <= 1200:
        return 'Premium'
    else:
        return 'Luxury'

df['price_category']=df["approx_cost(for two people)"].apply(price_bucket)
        
```

```python
# verifying column price_category

print(df['price_category'].value_counts())
```

```python
df["rate"].value_counts()
```

```python
def rating_label(rating):
    if rating >= 4.5:
        return "Excelent"
    elif rating >= 4.0:
        return "Very Good"
    elif rating >= 3.5:
        return "Good"
    elif rating >= 3.0:
        return 'Average'
    else:
        return 'Poor'

df["rating_label"] = df["rate"].apply(rating_label)
        
```

```python
# verifying column rating_label
print(df["rating_label"].value_counts())
```

```python
# Their's a problem in Cuisine column:
# "North Indian, Chinese, Continental"
#  we have to get only first cuisine so that we can organize different cuisines specifiacally into new column primary_cuisine

df['primary_cuisine'] = (
    df['cuisines']
    .str.split(',')   
    .str[0]           
    .str.strip()      
)

# Verifying column
print(df['primary_cuisine'].value_counts())
```
